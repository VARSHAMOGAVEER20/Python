```python
#Multiple Inheritance
class Father:
    def father_info(self):
        print("father:Hardworking")
class Mother:
    def mother_info(self):
        print("mother:caring")
class child(Father,Mother):
    def child_info(self):
        print("Child:Smart")
obj=child()
obj.father_info() 
obj.mother_info()
obj.child_info()               
        


```
