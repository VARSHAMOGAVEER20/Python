```python
class parent:
    def show(self):
        print("This is parent class")
class child1(parent):
    def display1(self):
        print("This is child1 class")
class child2(parent):
    def display2(self):
        print("This is child2 class")
obj1=child1() 
obj2=child2()
obj1.show()
obj1.display1()
obj2.show()
obj2.display2()
              
        
```
