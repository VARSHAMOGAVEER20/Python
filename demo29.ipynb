```python

#Multilevel inheritance
class GrandFather:
    def feature1(self):
        print("feature1")
class parent(GrandFather):
    def feature2(self):
        print("feature2")
class child(parent):
    def feature3(self):
        print("feature3")
obj=child()
obj.feature1() 
obj.feature2()
obj.feature3()               
        
```
