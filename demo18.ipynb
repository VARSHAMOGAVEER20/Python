```python
#new dictionary
student={"name":"Varsha","age":20,"course":"BCA"}
print(student)

#Add new pair
student["grade"]="A"
print(student)

#update value
student["age"]=22
print(student)

#remove by key
student.pop("course")
print(student)

#remove last item
student.popitem("age")
print(student)

#keys only
student.keys()
print(student)

#values only
student.values()
print(student)

#Items
student.items()
print(student)

#clear
student.clear()
print(student)
```
