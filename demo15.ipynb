List and its operation

```python

#New list
fruits=["Apple","Banana","Grapes"]
print(fruits[1])

#update
fruits[1]="Mango"
print(fruits)

#add
fruits.append("Orange")
print(fruits)

#remove by value
fruits.remove("Apple")
print(fruits)

#remove by index
fruits.pop(0)
print(fruits)

#sort
fruits.sort()
print(fruits)

#reverse
fruits.reverse()
print(fruits)

#length
len(fruits)

#slicing
fruits[1:3]
print(fruits)

#clear
fruits.clear()
print(fruits)
```
